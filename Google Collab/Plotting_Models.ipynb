```python
# Step 1: Import Necessary Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the CSV Data
from google.colab import files
uploaded = files.upload()

# Load the CSV file into a DataFrame, and specify the delimiter
df = pd.read_csv(next(iter(uploaded)), skipinitialspace=True)

# Step 3: Check the DataFrame's columns to confirm header names
print(df.columns)

# Step 4: Process the Data
# Extracting the relevant columns and stripping any whitespace from the headers
df.columns = df.columns.str.strip()  # Remove any leading/trailing spaces
epochs = df['epoch']
train_box_loss = df['train/box_loss']
val_box_loss = df['val/box_loss']
train_cls_loss = df['train/cls_loss']
val_cls_loss = df['val/cls_loss']

# Step 5: Plot the Data
plt.figure(figsize=(14, 6))

# Add a main title
plt.suptitle('(d). Model 4', fontsize=20)

# Plot Box Loss
plt.subplot(1, 2, 1)
plt.plot(epochs, train_box_loss, label='Training Box Loss', color='blue')
plt.plot(epochs, val_box_loss, label='Validation Box Loss', color='orange')
plt.title('Box Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Box Loss')
plt.legend()
plt.grid()

# Plot Class Loss
plt.subplot(1, 2, 2)
plt.plot(epochs, train_cls_loss, label='Training Class Loss', color='blue')
plt.plot(epochs, val_cls_loss, label='Validation Class Loss', color='orange')
plt.title('Class Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Class Loss')
plt.legend()
plt.grid()

plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to make room for the main title
plt.show()
```
